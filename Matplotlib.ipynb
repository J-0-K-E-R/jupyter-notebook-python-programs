# Matplotlib

Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general
Matplotlib allows you to create reproducible figures programmatically.
http://matplotlib.org/

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(8, 14, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Functional Method

```python
plt.plot(x, y, 'p-')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('Title')
# plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y , 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'p-')
```

## O.O. Method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'rp-')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y, 'rp-')
axes1.set_title('Larger Plot')
axes2.plot(y, x, 'bp-')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x, y, 'rp-')
axes[0].set_title('Larger Plot')
axes[1].plot(y, x, 'bp-')
axes[1].set_title('Smaller Plot')

plt.tight_layout()
```

## Figure size and DPI

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (8, 2), dpi = 500)
axes[0].plot(x, y, 'rp-', label = 'X vs X-Squared')
axes[0].set_title('Larger Plot')
axes[1].plot(y, x, 'bp-', label = 'X vs Root-X')
axes[1].set_title('Smaller Plot')

axes[0].legend()
axes[1].legend()
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 2), dpi = 500)
axes.plot(x, y, 'rp-', label = 'X vs X-Squared')
axes.plot(x, x ** 3, 'bp-', label = 'X vs X-Cubed')

axes.legend(loc = 0) # location code (0 is for best location) or specify loaction (0.1, 0.1))
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', linewidth = 10, alpha = 0.8, linestyle = '--')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', lw = 10, ls = 'steps')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

Different Line Styles

```python
fig = plt.figure(dpi = 200)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', lw = 0.5, marker = '*')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', lw = 0.5, marker = 2, markersize = 10)
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', lw = 0.5, marker = 'o', markersize = 10, markerfacecolor = '#fbc531')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Control over axis appearance

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color = '#9c88ff', lw = 0.5, marker = 'o', markersize = 10, markerfacecolor = '#fbc531')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
axes.set_xlim([0, 20])
axes.set_ylim([0, 250])
```
