# This is an implementation of a model using a random mathematical formula.

```python
import numpy as np
import math
from sklearn.utils import shuffle
```

```python
data = np.arange(0,1000)
z = np.zeros(1000)
```

```python
training_data = data[:700]
```

```python
training_data
```

```python
max1 = training_data.max()
min1 = training_data.min()
max1, min1
```

```python
training_data = shuffle(training_data, random_state=0)
```

```python
training_data
```

```python
v = np.zeros(700)
for i in range(0,700) :
    y = (math.sqrt(max1 - training_data[i]) / 2)
    v[i] = y
```

```python
v
```

```python
model = [training_data, v]
```

```python
testing_data = data[701:]
```

```python
testing_data
```

```python
testing_data = shuffle(testing_data, random_state=0)
```

```python
testing_data
```

```python
c= [np.zeros(700), np.zeros(700)]
rt = np.zeros(300)
z = np.zeros(300)
```

```python
for i in testing_data:
    index = 0
    sum = 0
    for j in training_data:
        c[0][index] = math.sqrt((i**2) - (j**2))
        c[1][index] = v[index]
        index += 1
    z = sorted(c[0])
    for t in z[-6:-1]:
        loc = np.where(c[0] == t)
        sum += c[1][loc]
    rt[list(testing_data).index(i)] = sum/5
```

```python
result = [testing_data, rt]
```
