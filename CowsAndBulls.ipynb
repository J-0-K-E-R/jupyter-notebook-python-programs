```python
import random
cows = 0
bulls = 0
```

```python
def validateNumbers(userN, randN):
    for x in range(0,4):
        if userN[x] == randN[x]:
            cows += 1
        elif userN[x] in randN:
            bulls += 1
    print('{}: {} cows, {} bulls'.format(userN,cows,bulls))
```

```python
if __name__ == '__main__':
    the_switch = 1
    while (the_switch):
        flag = 1
        random_number = random.randint(1000,10000)
        while (flag):
            print("Enter a four digit number or enter 'exit' to exit :")
            user_input = input();
            if user_input == 'exit':
                flag = 0
                the_switch = 0
                continue
            elif len(user_input) == 4:
                if int(user_input) == random_number:
                    print("Great, you've guessed the right number!!!")
                    print("Cows:" + str(cows) + " Bulls:", bulls)
                    flag = 0
                else:
                    validateNumbers(user_input, str(random_number))
            else:
                print("Please enter a vaid value")
```
